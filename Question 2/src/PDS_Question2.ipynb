```python
import pandas as pd
```

```python
file_path = '/content/StudentsPerformance.csv'

# Data Loading using Pandas
df = pd.read_csv(file_path)

# Data Set Preview
df.head(10)
```

####Histogram representation of reading scores

```python
import matplotlib.pyplot as plt

plt.hist(df['reading score'], bins=10, edgecolor='k')
plt.title('Histogram of Reading Scores')
plt.xlabel('Reading Score')
plt.show()
```

###Scatterplot representation between Math Score and Reading Score

```python
import seaborn as sns
```

```python
print("Scatter Plot: Math Score vs Reading Score")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=df)
plt.title('Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

###Lineplot representation for Trend of Scores

```python
print("Line Plot: Trend of Scores")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df[['math score', 'reading score', 'writing score']], dashes=False)
plt.title('Trend of Scores')
plt.xlabel('Index')
plt.ylabel('Score')
plt.show()
```

###Q-Q Plot for math score

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# Select a numerical variable for the Q-Q plot
variable = 'math score'

# Create a Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(df[variable], dist="norm", plot=plt)
plt.title("Q-Q Plot for " + variable)
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Ordered Values")
plt.grid(True)
plt.show()
```

###Pie Chart representation for gender distribution

```python
print("Pie Chart: Gender Distribution")
plt.figure(figsize=(8, 8))
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```
